```python
import json
import re 
contents = []
for eaLine in open('cmsDataSchema.json', 'r'):
    data = json.loads(eaLine)
    contents.append(data)
```

```python
def parseDataType(dataType):
    columnVals = dataType['data_type']
    if columnVals.startswith('STRUCT'):
        result = re.search('<(.*)>', columnVals)
        return [ea for ea in result.group(1).split(',')]
    else:
        columnVals = dataType['column_name']
        return [columnVals]
```

```python
columns = [  parseDataType(ea) for ea in contents ]
```

```python
colFinal = [each for eaCol in columns for each in eaCol ]
```

```python
colList = [eaCol.split()[0] for eaCol in colFinal]
```
